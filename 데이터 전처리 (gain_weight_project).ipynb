### NaN값 처리하기  (컬럼의 특성마다 처리하는 방식이 달라짐)
### ex) 시간의 흐름과 관련이 있느냐 없느냐
### 여기서는 체중(아침), 체중(저녁)이 시간의 흐름과 관련이 있음.
### ----------------------------------------------------------------------------------


#### 1. 밀가루:  (1) 평균값을 넣는다. (2) NaN으로 그냥 둔다.
#### 2. 식사, 간식 : (1) 평균값을 넣는다. (2) NaN으로 그냥 둔다.
#### 3. 초록, 노랑, 빨강, 칼로리 : (1) 예측할 수 없다. NaN으로 그냥 둔다.
#### 4. 체중(아침), 체중(저녁) : (1) 평균값을 넣는다 (2) 다음 날 혹은 전 날의 체중 값을 넣는다.
#### 5. 장활동 : (1) 예측할 수 없다. NaN으로 그냥 둔다. 

### NaN 값 처리하기 
### 1. 밀가루 (평균 값 넣기)

```python
## 밀가루 컬럼에 NaN이 있는지 확인하기

data[data['밀가루'].isnull()] # data의 밀가루 컬럼이 누락된 data를 색인 

# 한 개의 행이 색인되었다. 
```

```python
# 이 행의 거의 모든 값이 NaN이지만 체중(아침)값이 기록되어있어 제거하지는 않겠다. 

# 밀가루 평균 값
data['밀가루'].mean()
```

```python
data['밀가루'] = data['밀가루'].fillna(496) 

# 밀가루 컬럼의 누락된 셀에 평균값을 넣는다.
# 마찬가지로 식을 쓰고 난 후 컬럼에 담아야 적용이 된다. (data['밀가루'] = )

data[data['밀가루'].isnull()] 

# 밀가루 컬럼에서 누락된 값이 색인되지 않는다. 
```

### 2. 물 (평균 값 넣기)

```python
data[data['물'].isnull()]
```

```python
data['물'].mean()
```

```python
data['물'] = data['물'].fillna(904)
```

### 3.식사, 간식 횟수 (평균 값 넣기)

```python
data['식사'].mean()
```

```python
data['간식'].mean()
```

```python
# 식사에는 2번 (2.2번은 없으므로), 간식에는 4번을 넣겠다.

data['식사'] = data['식사'].fillna(2)
data['간식'] = data['간식'].fillna(4)
```

### 4. 체중 (전 날의 체중값 넣기) 
#### 체중은 시간의 흐름에 따라 점진적으로 변하는 것이다. 
#### 단순히 평균 값을 넣기 보다 전 날의 값을 가져와 넣는 것이 더 정확할 듯 하다. 

### 4-1 체중(아침)값 넣기

```python
data[data['체중(아침)'].isnull()] # 누락된(nan) 체중(아침) 데이터 색인
```

```python
data['체중(아침)'].fillna(data['체중(아침)'].shift(+1))   

# shift(+1)은 전 행의 값을 가져온다. 
# shift(-1)은 다음 행의 값을 가져온다. 
# 10행을 보면 9행의 값인 39.4를 가져와 채운 것을 알 수 있다. 
```

```python
data['체중(아침)'] = data['체중(아침)'].fillna(data['체중(아침)'].shift(+1)) 

data[data['체중(아침)'].isnull()] # naa 값이 색인되지 않는 것으로 보아 다 채워진 듯 하다. 
```

### 4-2 체중(저녁)값 넣기

```python
# 똑같은 방법으로 체중(저녁)을 채운다. 

data[data['체중(저녁)'].isnull()] # nan 값의 데이터 색인 먼저하고
```

```python
data['체중(저녁)'] = data['체중(저녁)'].fillna(data['체중(저녁)'].shift(+1))  
data['체중(저녁)'] = data['체중(저녁)'].fillna(data['체중(저녁)'].shift(+1)) 

# nan값 중 연속된 행이 있어 두 번 실행했다. (9행과10행, 34행과35행)

data[data['체중(저녁)'].isnull()] # nan값 색인 

# 0행은 가져올 이전 행이 없어 값이 채워지지 않았다.
```

```python
data['체중(저녁)'] = data['체중(저녁)'].fillna(data['체중(저녁)'].shift(-1))

# 다음 행의 값을 가져와 값을 채운다.

data[data['체중(저녁)'].isnull()] # nan값이 이제 없다. 잘 채워졌음. 
```

## 장활동 전처리하기
### 장활동(clean)이라는 새로운 컬럼을 만들어 오직 good/soso/bad만 나오도록 바꾼다.

```python
data[data['장활동'].str.contains('bad')==True].tail(7)

# 새로운 컬럼으로 만드는 이유는 
# bad옆에 다양한 이유로 값이 들어가있는데 이렇게 되면 value에 통일성이 없어 분석하기 불편하다. 

# bad가 들어가면 무조건 bad만 나타나도록 새로운 컬럼을 만들겠다.
```

```python
def 장활동_clean(장활동):        #'장활동_clean'이라는 함수를 새로 만들고 그 함수 안에는 '장활동'이란 인자가 있다.  
    if pd.isnull(장활동):        # 만약 장활동에 nan값이 있으면 그 값을 그대로 반환해라
        return 장활동
    if 'good' in 장활동:         # 만약 장활동에 good이란 단어가 있으면 good을 반환해라
        return 'good'
    if 'soso' in 장활동:         # 만약 장활동에 soso란 단어가 있으면 soso를 반환해라 
        return 'soso'
    if 'bad' in 장활동:          # 만약 장활동에 bad란 단어가 있으면 bad를 반환해라 
        return 'bad'
```

```python
data['장활동(clean)'] = data['장활동'].apply(장활동_clean) 

# '장활동'이란 인자는 data['장활동']에 들어있는 값들 하나하나가 된다. 
# data['장활동']에 '장활동_clean' 함수를 적용(apply)하면 
# data['장활동']컬럼의 행 하나하나를 돌면서 함수를 처리한다. 
# 그 처리된 값을 data['장활동(clean)']이라는 새로운 컬럼에 넣는다.
```

```python
data.loc[data['장활동'].str.contains("bad")==True, ['장활동','장활동(clean)']]

# 장활동 컬럼에서 bad가 들어가있는 것만 골라내 그것에 해당하는 장활동 컬럼과 장활동(clean)컬럼을 색인해라.

# 이쁘게 bad만 걸러져 나왔음을 알 수 있다. 
```
